**Importing Numpy Library**

```python
import numpy as np
```

**Matrix Addition**

Note: Two Matrices can only be added if they have the same shape.

```python
a = np.array([[2,3],[4,5]])

b = np.array([[6,7],[8,9]])

print(a+b)
```

```python
# creating two matrices with random values

matrix_1 = np.random.randint(10, size=(3,3))

matrix_2 = np.random.randint(30, size=(3,3))



print(matrix_1)

```

```python
print(matrix_2)
```

```python
matrix_addition = np.add(matrix_1, matrix_2)

print(matrix_addition)
```

**Matrix Subtraction**

Note: Two Matrices can only be subtracted if they have the same shape.

```python
a = np.array([[6,7],[8,9]])

b = np.array([[1,2],[3,4]])

print(a-b)
```

```python
matrix_3 = np.random.randint(10, size=(3,3))

matrix_4 = np.random.randint(30, size=(3,3))



print(matrix_3)
```

```python
print(matrix_4)
```

```python
matrix_subtraction = np.subtract(matrix_3, matrix_4)

print(matrix_subtraction)
```

**Multiplying a matrix by a scalar**

```python
x = 5

y = np.random.randint(10, size=(3,3))

print(y)
```

```python
product = x * y

print(product)
```

**Multiplying two Matrices**

Note: The number of columns in the first matrix should be equal to the number of rows in the second matrix.

```python
matrix_5 = np.random.randint(10, size=(3,3))

matrix_6 = np.random.randint(30, size=(3,4))



print(matrix_5)
```

```python
print(matrix_6)
```

```python
product = np.dot(matrix_5, matrix_6)

print(product)
```

**Element wise Matrix Multiplication**

```python
matrix_7 = np.random.randint(10, size=(3,3))

matrix_8 = np.random.randint(10, size=(3,3))


print(matrix_7)
```

```python
print(matrix_8)
```

```python
# Both matrix should have same shape for np.multiply
product = np.multiply(matrix_7, matrix_8)

print(product)
```
