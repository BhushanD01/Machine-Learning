```python
# importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

**GridSearchCV**

GridSearchCV is used for determining the best parameters for our model

```python
model = SVC()
```

```python
# hyperparameters
parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
# grid search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to our model
classifier.fit(X, Y)
```

```python
# Highest Accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# Loading the results to Pandas Dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
grid_search_result
```

**RandomizedSearchCV**

```python
model = SVC()
```

```python
parameters = {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X, Y)
```

```python
classifier.cv_results_
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```
