Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sns.set improves the aesthetics of the plots by modifying the underlying Matplotlib settings
sns.set()
```

Plotting a Vector

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color="b")
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Plotting Multiple Vectors

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color="b")
plt.quiver(0,0,-3,-8,scale_units='xy',angles='xy',scale=1,color="r")
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Addition of two Vectors

```python
vector_1 = np.array([4,5])
vector_2 = np.array([-3,-8])

addition = vector_1 + vector_2

print(addition)
```

Subtraction of two Vectors

```python
vector_1 = np.array([4,5])
vector_2 = np.array([-3,-8])

diff = vector_1 - vector_2

print(diff)
```

Multiplying a vector by a scalar

```python
vector_3 = np.array([-9,-6])

prod = vector_3 * 2

print(prod)
```

```python
vector_4 = np.array([-3,7])

prod = vector_4 * -0.5

print(prod)
```

Dot Product of two Vectors

```python
a = np.array([2,3])

b = np.array([1,5])

dot_prod = np.dot(a,b)

print(dot_prod)
```

Cross Product of two Vectors

```python
c = np.array([15,20,25])

d = np.array([5,10,15])

cross_prod = np.cross(c,d)

print(cross_prod)
```

Projection of "u" vector on "v" vector

parallelprojection1.jpg

```python
u = np.array([25,42,57])

v = np.array([67,82,12])

# magnitude of "v" vector

mag_v = np.sqrt(np.sum(v**2))

projection_u_on_v = (u.dot(v)/mag_v**2)*v

print("Projection of u vector on v vector: ",projection_u_on_v)

```
